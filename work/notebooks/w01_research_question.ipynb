## Section 2: The question: decision, action, cost of a wrong call

Decision: Deciding which high-value URLs require immediate content updates or technical SEO re-optimization within the next 2–4 weeks.

Who acts on it: SEO Managers and Content Strategy leads.

Cost of False Positive (Flagged healthy page as decaying): Wastes 2–4 hours of content writer effort updating a page that was performing normally.

Cost of False Negative (Missed a genuinely decaying page): Page drops out of top search positions, resulting in a 40–60% loss in organic leads that takes months to recover.

```python
# Decision impact cost estimation matrix
cost_false_positive_hours = 3.0
cost_false_negative_traffic_loss_pct = 0.50

print(f"Decision impact model initialized:")
print(f"- False Positive Cost: ~{cost_false_positive_hours} hours of wasted manual review")
print(f"- False Negative Cost: ~{cost_false_negative_traffic_loss_pct * 100}% organic traffic drop")
```
